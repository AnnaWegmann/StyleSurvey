```python
# needs Artifacts/anthology+abstracts.bib.gz, download from https://aclanthology.org/anthology+abstracts.bib.gz

import gzip
import bibtexparser
import os
bibfile = "../Artifacts/anthology+abstracts.bib.gz"
# check if file exists
if not os.path.exists(bibfile):
    raise FileNotFoundError(f"Bib file not found: {bibfile}")
with gzip.open(bibfile, "rt", encoding="utf-8") as f:
    bib_database = bibtexparser.load(f)
```

```python
# largely generated with GPT 5
import re
from collections import Counter

# Whole-word, case-insensitive "style"
STYLE = re.compile(r"\bstyle\b", re.IGNORECASE)

# Simple TeX cleanup so matching is robust
def clean_text(s: str) -> str:
    s = re.sub(r"\\[a-zA-Z]+(\{[^}]*\})?", " ", s)  # remove \commands (crudely)
    s = s.replace("{", " ").replace("}", " ")
    s = re.sub(r"\s+", " ", s).strip()
    return s

# Count of *papers* per year that contain "style"
papers_with_style_per_year = Counter()

for entry in bib_database.entries:
    year = entry.get("year", "").strip()
    if not re.fullmatch(r"\d{4}", year):
        continue  # skip malformed/missing years

    text = clean_text(entry.get("title", "") + " " + entry.get("abstract", ""))
    if STYLE.search(text):
        papers_with_style_per_year[year] += 1

# Print nicely, sorted by year
for y in sorted(papers_with_style_per_year.keys()):
    print(f"{y}: {papers_with_style_per_year[y]}")
```

### Output ACL Anthology papers mentioning "style" and "measur"

```python
# largely generated with GPT 4o

results = []
for entry in bib_database.entries:
    title = entry.get("title", "")
    abstract = entry.get("abstract", "")
    url = entry.get("url", "")
    year = entry.get("year", "")

    content = f"{title} {abstract}".lower()
    if "style" in content and "measur" in content:
        results.append((title, url, year))

# Print results
print(f"Found {len(results)} matching papers.\n")
for title, url, year in results:
    print(f"{title} ({year})\n{url}\n")
```

```python
# generated GPT 4o
import gzip
import re

bibfile = "../../Artifacts (too large to sync)/anthology+abstracts.bib.gz"

def extract_field(entry: str, field: str) -> str:
    # Match field = {multi-line content}
    pattern = rf"{field}\s*=\s*{{(.*?)}}"
    match = re.search(pattern, entry, re.IGNORECASE | re.DOTALL)
    if match:
        return match.group(1).replace("\n", " ").strip()
    # Fallback: match field = "..."
    pattern = rf'{field}\s*=\s*"([^"]+)"'
    match = re.search(pattern, entry, re.IGNORECASE)
    if match:
        return match.group(1).strip()
    return ""

with gzip.open(bibfile, "rt", encoding="utf-8") as f:
    entries = f.read().split("@")

results = []
for entry in entries:
    if not entry.strip():
        continue
    entry_text = "@" + entry

    title = extract_field(entry_text, "title")
    abstract = extract_field(entry_text, "abstract")
    url = extract_field(entry_text, "url")
    year = extract_field(entry_text, "year")

    # Only search in title and abstract (not booktitle)
    search_text = f"{title} {abstract}".lower()
    if "style" in search_text and "measur" in search_text:
        results.append((title, url, year, entry))

# Print results
print(f"Found {len(results)} papers mentioning both 'style' and 'measure' in title/abstract.\n")
for title, url, year, entry in results:
    print(f"{title} ({year})\n{url}\n{entry}\n") # 
```
